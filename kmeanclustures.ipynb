```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
print("Hello1")
```

```python
df=pd.read_csv("japan.csv")
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df["longitude"],df["latitude"])
```

```python
km=KMeans(n_clusters=2)
km
```

```python
value1=km.fit_predict(df[['longitude','latitude']])
value1
df['kmean']=value1
df.head()
df1=df[df.kmean==0]
df2=df[df.kmean==1]
plt.scatter(df1.longitude,df1['latitude'],color='Red')
plt.scatter(df2.longitude,df2['latitude'],color='Green')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="Purple",marker="*",label="cent")
```

```python
km.cluster_centers_
```

```python
km1=KMeans(n_clusters=3)
value2=km1.fit_predict(df[['longitude','latitude']])
value2
df['kmean']=value2

df1=df[df.kmean==0]
df2=df[df.kmean==1]
df3=df[df.kmean==2]

plt.scatter(df1.longitude,df1['latitude'],color='Orange')
plt.scatter(df2.longitude,df2['latitude'],color='Green')
plt.scatter(df3.longitude,df3['latitude'],color='Red')
plt.scatter(km1.cluster_centers_[:,0],km1.cluster_centers_[:,1],color="Purple",marker="*",label="cent")
```

```python
k_range=range(1,10)
sse=[]
for i in k_range:
  km=KMeans(n_clusters=i)
  km.fit(df[['longitude','latitude']])
  sse.append(km.inertia_)
  
```

```python
plt.xlabel("K")
plt.ylabel("Sum of squared error")
plt.plot(k_range,sse)
plt.grid(True)
```
